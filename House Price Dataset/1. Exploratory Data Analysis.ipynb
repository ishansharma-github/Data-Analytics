###  Missing values

```python
#Cheking the percentage of missing values in each feature

#finding features with null values
a = list(df.columns)
features_with_null_values = list(filter(lambda x: df[x].isnull().sum()>0 ,a))

#percentage of null values in each feature.
for i in features_with_null_values:
    print("Percentage of missing values in",i,":",np.round(df[i].isnull().mean()*100,4),"%")
```

>  Finding relationship between missing values and Sales price

```python
for i in features_with_null_values:
    c = df.copy()
    
    # let's make a variable that indicates 1 if the observation was missing or zero otherwise
    c[i] = np.where(c[i].isnull(), 1, 0)
    
    # let's calculate the mean SalePrice where the information is missing or present
    c.groupby(i)['SalePrice'].median().plot.bar()
    plt.title(i)
    plt.show()
```

Here, we can conclude that Null values are contributing to sales price hence we need to take care of null values

#### 1) Numerical Features

```python
numerical_features = list(filter(lambda x: df[x].dtype == 'int64' or df[x].dtype == 'float64', df.columns))
print("Numbers of Numerical Features",len(numerical_features))
```

```python
df[numerical_features]
```

#### 2) Date-Time variables

```python
dt_variables = list(filter(lambda x: ('Yr' in x) or ("Year" in x) ,df.columns))
dt_variables
```

```python
df[dt_variables].sort_values(dt_variables)
```

#### 3) Discrete numerical Features

```python
discrete_features = list(filter(lambda x: len(df[x].unique())<25 and (x not in dt_variables),numerical_features))
print("Number of discrete vaiables are:",len(discrete_features))
```

```python
for i in discrete_features:
    c = df.copy()
    c.groupby(df[i])['SalePrice'].median().plot.bar()
    plt.xlabel(i)
    plt.ylabel('SalePrice')
    plt.title(i)
    plt.show()
```

#### 4) Continuous numerical variables.

```python
continuous_features = list(filter(lambda x:x not in (dt_variables + discrete_features + ['Id'] ),numerical_features))
print("Number of continuous vaiables are:",len(continuous_features))
```

```python
for i in continuous_features:
    c = df.copy()
    c[i].hist(bins = 25)
    plt.xlabel(i)
    plt.ylabel("Sale Price") 
    plt.title(i)
    plt.show()
    
```

###### plotting in terms of logarithm for better insight.

```python
 for i in continuous_features:
        c = df.copy()
        if 0 in c[i].unique():
            pass
        else:
            c[i] = np.log(c[i])
            c['SalePrice'] = np.log(c['SalePrice'])
            plt.scatter(c[i],c['SalePrice'])
            plt.xlabel(i)
            plt.ylabel('SalesPrice')
            plt.title(i)
            plt.show()
```

### Outliers

```python
for i in continuous_features:
    c = df.copy()
    if 0 in c[i].unique():
        pass
    else:
        c[i] = np.log(c[i])
        c.boxplot(column = i)
        plt.xlabel(i)
        plt.title(i)
        plt.show()
```
