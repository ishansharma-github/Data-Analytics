### Missing values

1. Handling categorical features containing null values.

```python
Feature_with_na = list(filter(lambda x: (df[x].isnull().sum()>0) and (df[x].dtype == "O"),df.columns))
df[Feature_with_na]
```

```python
# percentage of null values in categorical features
for i in Feature_with_na:
    print("In feature",i,"percentage of null values is",round(df[i].isnull().mean()*100,2),"%")
```

```python
# Replacing na with word "missing".
data = df.copy()
```

```python
df[Feature_with_na] = data[Feature_with_na].fillna("Missing")
```

```python
df[Feature_with_na].isnull().sum()
```

2. Handling numerical features containing null values.

```python
# Numerical features with na
Feature_with_na = list(filter(lambda x: (df[x].isnull().sum()>0) and (df[x].dtype != "O"),df.columns))
df[Feature_with_na]
```

```python
# Percentage of na values in numerical features
for i in Feature_with_na: 
    print(i,":",round(df[i].isnull().mean()*100,2),"%")
```

```python
# Replacing missing values by median.
for i in Feature_with_na:
    median = df[i].median()
    
    # to remember where we we substituted nan with median
    df[i+"nan"] = np.where(df[i].isnull(),1,0)
    
    # replacing nan with median
    df[i].fillna(median,inplace = True)
```

```python
naornot = list(map(lambda x: x+"nan",Feature_with_na))
df[Feature_with_na+naornot].head(25)
```

3. Handling Temporal Variables

```python
for i in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
    df[i] = df['YrSold'] - df[i]
```

```python
df[['YearBuilt','YearRemodAdd','GarageYrBlt']].head()
```

## Numerical Variables

since the numerical variable are skewed we will perform log normal distribution.

```python
df.head()
```

```python
# Converting skewed destribution to log normal distribution
# Now we need those num vaiables which are not zero, we found this in EDA notebook.

features = ['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for i in features:
    df[i] = np.log(df[i])
```
